```python
#1 Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
#3 Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
fig, ax = plt.subplots()
ax.plot(x, y, label='Line Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Line Plot with Labels')
ax.legend()
plt.show()
```

```python
#4 Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
from io import StringIO
data_str = """1 2
2 4
3 6
4 8
5 10"""
data = [line.split() for line in data_str.split('\n')]
x = [float(row[0]) for row in data if row]
y = [float(row[1]) for row in data if row]
fig, ax = plt.subplots()
ax.plot(x, y, label='Line Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Line Plot with Labels')
ax.legend()
plt.show()
```

```python
#5 Write a Python program to plot two or more lines with legends, different widths and colors.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]
fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', linewidth=2, label='Line 1')
ax.plot(x, y2, color='red', linewidth=3, label='Line 2')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Multiple Lines with Legends')
ax.legend()
plt.show()
```

```python
#6 Write a Python program to plot quantities which have an x and y position.

import matplotlib.pyplot as plt
x_positions = [1, 2, 3, 4, 5]
y_positions = [2, 4, 6, 8, 10]
fig, ax = plt.subplots()
ax.scatter(x_positions, y_positions, color='blue', marker='o', label='Data Points')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Quantities with X and Y Positions')
ax.legend()
plt.show()
```

```python
#7 Create plot using subplots(), scatter, and some other methods on the Axes object:

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 50)
y1 = x + np.random.randn(50)
y2 = 2*x + np.random.randn(50)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
axes[0].scatter(x, y1, color='blue', marker='o', label='Scatter Plot 1')
axes[0].set_xlabel('X-axis Label')
axes[0].set_ylabel('Y-axis Label')
axes[0].set_title('Subplot 1')
axes[0].legend()
axes[1].scatter(x, y2, color='red', marker='x', label='Scatter Plot 2')
axes[1].set_xlabel('X-axis Label')
axes[1].set_ylabel('Y-axis Label')
axes[1].set_title('Subplot 2')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
#8 Generate random data for x and y and create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
num_points = 50
x = np.random.rand(num_points) * 10
y = x + np.random.randn(num_points)
plt.scatter(x, y, color='blue', marker='o', label='Random Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot with Random Data')
plt.legend()
plt.show()
```

```python
#9 Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.

import matplotlib.pyplot as plt
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 500, 200, 600]
plt.bar(products, sales, color='red')
plt.xlabel('Products')
plt.ylabel('Sales (in units)')
plt.title('Product Sales Comparison')
plt.show()
```

```python
#10 Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randn(1000)
plt.hist(data, bins=20, color='red', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
#11 Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

import matplotlib.pyplot as plt
categories = ['Housing', 'Utilities', 'Groceries', 'Transportation', 'Entertainment']
expenses = [30, 15, 25, 20, 10]
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['red','lightgreen','yellow','skyblue','magenta'])
plt.title('Expense Distribution in Budget')
plt.show()
```
